```python
hierarchy = {"background":"", "C0":{"C2":{"C7":{"C8":{"C9":""}}}, "C3":"", "C4":""}, "C1":{"C5":"", "C6":""}}


# takes the hierarchy structure dictionary and returns the maximum depth and the total number of classes for each depth
def get_level_classes(hierarchy, depth=0, result=None):
    if result is None:
        result = {}
    if not isinstance(hierarchy, dict) or not hierarchy:
        return result
    for key, value in hierarchy.items():
        if depth not in result:
            result[depth] = []
        result[depth].append(key)
        if isinstance(value, dict):
            get_level_classes(value, depth + 1, result)
    return result

level_clases = [x for x in get_level_classes(hierarchy).values()]
depth = len(level_clases)


print(depth)
print(level_clases)





# n_layers = read_hierarchy(hierarchy)
```
