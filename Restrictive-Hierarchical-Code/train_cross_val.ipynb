## Cross val Eval

```python
import subprocess
import numpy as np

# List to store results from each fold
results = []

# Loop for 5 folds
for fold in range(1, 6):
    # Modify arguments for the current fold
    fold_args = base_args.copy()
    fold_args[1] = f"F:/Github Repos/Restrictive-Hierarchical-Code/runs/exp1/fold_{fold}"  # Update save-path
    fold_args[3] = f"I:/Datasets/Hierarchical Datasets/TL_pano/processed/split/train/fold_{fold}"  # Update data-root
    fold_args[5] = f"I:/Datasets/Hierarchical Datasets/TL_pano/processed/split/val/fold_{fold}"  # Update val-dataset

    # Run predictEval.py with the modified arguments
    result = subprocess.run(["python", "predictEval.py"] + fold_args, capture_output=True, text=True)

    # Extract the result (assuming the result is a single numeric value in stdout)
    try:
        result_value = float(result.stdout.strip())
        results.append(result_value)
    except ValueError:
        print(f"Error parsing result for fold {fold}: {result.stdout}")

# Calculate mean and standard deviation
if results:
    mean_result = np.mean(results)
    std_result = np.std(results)
    print(f"Mean Result: {mean_result}")
    print(f"Standard Deviation: {std_result}")
else:
    print("No valid results to calculate statistics.")
```
