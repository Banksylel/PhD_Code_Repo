### run this code to count the ground truth data

```python
import pandas as pd
import ast
from collections import defaultdict, Counter

# Read the CSV file
df = pd.read_csv('F:/Github Repos/study/study_code/processed_data/task_1_inferred_ground_truth.csv')

# Initialize counters
key_counts = Counter()
value_counts = defaultdict(Counter)

# Traverse each row
for index, row in df.iterrows():
    # Literal eval the dictionary in 'tooth_annotations' column
    ann = ast.literal_eval(row['tooth_annotations'])
    # Count keys and values
    # print(row)
    # print(ann)
    for item in ann:
        for key, value in item.items():
            key_counts[key] += 1
            value_counts[key][value] += 1

# Print the results
print("Key counts:")
for key, count in key_counts.items():
    print(f"{key}: {count}")

print("\nValue counts per key:")
for key, v_counts in value_counts.items():
    print(f"{key}: {dict(v_counts)}")
```
